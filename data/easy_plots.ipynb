```python
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option("max_rows", 500)
```

```python
df = pd.read_csv('df_angled')
df_1 = pd.read_csv('df_horizontal')
df_2 = pd.read_csv('df_benchmark')
df_3 = pd.read_csv('df_single')
df_4 = pd.read_csv('df_pillars')
```

```python

df=df.loc[df['avrg_velocity']<25]
df_1=df_1.loc[df_1['avrg_velocity']<25]
df_2=df_2.loc[df_2['avrg_velocity']<25]
df_3=df_3.loc[df_3['avrg_velocity']<25]
df_4=df_4.loc[df_4['avrg_velocity']<25]

df['win']=df['succes']/df['peop_num']
df_1['win']=df_1['succes']/df_1['peop_num']
df_2['win']=df_2['succes']/df_2['peop_num']
df_3['win']=df_3['succes']/df_3['peop_num']
df_4['win']=df_4['succes']/df_4['peop_num']


```

```python
df.head(4)
```

```python
df.plot('peop_num','avrg_time',kind='line')
```

```python
selector='win'

angle=df.groupby('peop_num')[selector].mean()
horizon=df_1.groupby('peop_num')[selector].mean()
bemchmark=df_2.groupby('peop_num')[selector].mean()
single=df_3.groupby('peop_num')[selector].mean()
plt.plot(angle,color='pink',label='angle')
plt.plot(horizon,label='horizon')
plt.plot(bemchmark,label='benchmark')
plt.plot(single,label='single')
plt.legend()
```

```python
selector='avrg_time'

angle=df.groupby('peop_num')[selector].mean()
horizon=df_1.groupby('peop_num')[selector].mean()
bemchmark=df_2.groupby('peop_num')[selector].mean()
single=df_3.groupby('peop_num')[selector].mean()
pillars=df_4.groupby('peop_num')[selector].mean()


plt.plot(angle,color='pink',label='double diagonal')
plt.plot(horizon,label='horizontal')
plt.plot(bemchmark,label='benchmark')
plt.plot(single,label='single diagonal')
plt.plot(pillars,label='pillars')


plt.title('average time for an agent to traverse the system')
plt.xlabel('number of agents')
plt.ylabel('number of time steps')
plt.legend()
```
