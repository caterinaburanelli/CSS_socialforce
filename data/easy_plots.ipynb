```python
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option("max_rows", 500)
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('complete_fixed')
```

```python
sns.lineplot(data=df.loc[df['layout']=='benchmark'], x="peop_num", y='avrg_time',label='benchmark')
sns.lineplot(data=df.loc[df['layout']=='single'], x="peop_num", y='avrg_time',label='single')
sns.lineplot(data=df.loc[df['layout']=='pillars'], x="peop_num", y='avrg_time',label='pillars')
sns.lineplot(data=df.loc[df['layout']=='horizontal'], x="peop_num", y='avrg_time',label='horizontal')
sns.lineplot(data=df.loc[df['layout']=='angled'], x="peop_num", y='avrg_time',label='angled')
plt.xlabel('number of agents in the system')
plt.ylabel('average time to traverse the system')
plt.title('average time')
plt.legend()


```
