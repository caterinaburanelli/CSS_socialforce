```python
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option("max_rows", 500)
import numpy as np
import seaborn as sns
```

```python
df_1= pd.read_csv('df_single')
df_2= pd.read_csv('df_benchmark')
df_3= pd.read_csv('df_angled')
```

```python
sns.lineplot(data=df_2.loc[df_2['layout']=='benchmark'], x="peop_num", y='avrg_velocity',label='benchmark')
sns.lineplot(data=df_1.loc[df_1['layout']=='single'], x="peop_num", y='avrg_velocity',label='single')
sns.lineplot(data=df_3.loc[df_3['layout']=='angled'], x="peop_num", y='avrg_velocity',label='angled')
plt.xlabel('number of agents in the system')
plt.ylabel('average velocity')
plt.title('average velocity')
plt.legend()
```

```python
sns.lineplot(data=df_2.loc[df_2['layout']=='benchmark'], x="peop_num", y='avrg_time',label='benchmark')
sns.lineplot(data=df_1.loc[df_1['layout']=='single'], x="peop_num", y='avrg_time',label='single')
sns.lineplot(data=df_3.loc[df_3['layout']=='angled'], x="peop_num", y='avrg_time',label='angled')
plt.xlabel('number of agents in the system')
plt.ylabel('atime')
plt.title('time')
plt.legend()
```
